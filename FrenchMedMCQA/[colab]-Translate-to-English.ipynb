```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='fr', target='en')
```

```python
import pandas as pd
train_df = dataset = pd.read_json("./data/train.json")
```

```python
train_df
```

```python
train_df.iloc[0]
```

```python
translator.translate(train_df.iloc[0]['question'])
```

```python
from tqdm import tqdm

train_en_data = []
for index, row in tqdm(train_df.iterrows()):
    question = translator.translate(row["question"])
    answers = row["answers"]
    new_answers = {}
    for choice, answer in answers.items():
        new_answers[choice] = translator.translate(answer)
    data = {
        'id': row["id"],
        'question': question,
        'answers': new_answers,
        'correct_answers': row["correct_answers"],
        'subject_name': row["subject_name"],
        'nbr_correct_answers': row["nbr_correct_answers"]
    }
    train_en_data.append(data)
```

```python
train_en_df = pd.DataFrame(train_en_data)
```

```python
train_en_df
```

```python
train_en_df.to_csv('./data/train_en.csv', index=False)
```
